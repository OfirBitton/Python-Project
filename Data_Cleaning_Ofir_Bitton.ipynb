#Data Cleaning

```python
import pandas as pd
import numpy as np
```

```python
democracy_df = pd.read_excel('/content/ENGLISH_VERSION_IsraelDemocracySurvey2021_with_changes.xlsx')
```

###Removing Columns

```python
democracy_df = democracy_df.drop(['Unnamed: 0.1','Unnamed: 0.2','Unnamed: 1','w.Jew_Arab','w.Sex','w.Reli_Jew','w.Age','Reli_Arab','W.TOTAL'],axis=1)
democracy_df.head(10)
```

###Handling Nulls

####part1

```python
for row in democracy_df.index:
  if pd.isnull(democracy_df.loc[row,'(Arabs) Are you:']):
    if democracy_df.loc[row,'sector'] == 'Jewish':
      democracy_df.loc[row,'(Arabs) Are you:'] = 'No Arabic'
    else:
      democracy_df.loc[row,'(Arabs) Are you:'] = 'Other'

democracy_df.head(10)
```

```python
democracy_df['(Arabs) Are you:']
```

```python
for row in democracy_df.index:
  if pd.isnull(democracy_df.loc[row,'(To Jews) How do you religiously define yourself?']):
    if democracy_df.loc[row,'sector'] == 'Arabic':
      democracy_df.loc[row,'(To Jews) How do you religiously define yourself?'] = 'No Jewish'
    else:
      democracy_df.loc[row,'(To Jews) How do you religiously define yourself?'] = 'Other'

democracy_df.head(10)
```

```python
democracy_df['(To Jews) How do you religiously define yourself?']
```

####part2

```python
democracy_df['What are you?'].isnull().sum()
```

```python
democracy_df['What are you?'].value_counts()
```

```python
groups_mean = democracy_df.groupby(by='What are you?')['How old are you?'].mean()
groups_mean
```

```python
for row in democracy_df.index:
  if pd.isnull(democracy_df.loc[row,'How old are you?']):
    if democracy_df.loc[row,'What are you?'] == 'Man':
      democracy_df.loc[row,'How old are you?'] = groups_mean['Man']
    else:
      democracy_df.loc[row,'How old are you?'] = groups_mean['Woman']

democracy_df
```

##Categorical/Binary/Ordinal conversion

```python
converted_df = democracy_df.copy()
converted_df.head(10)
```

####Categorical columns - dummy variables



```python
categorical_columns = ['What stronger tension today in Israeli society? (Presented in rotation)','(Arabs) Are you:',
                      'Which party vote in the last election for the Knesset (March 2021):', 'district']
converted_df = pd.get_dummies(data=converted_df, columns=categorical_columns)
```

```python
converted_df.head(10)
```

####convrt Binary to 0/1

```python
binary_dict = {'sector' : {'Arabic' : 0, 'Jewish' : 1}, \
               'source' : {'Internet' : 0, 'Telephone' : 1}, \
               'What are you?' : {'Woman' : 0, 'Man' : 1}}
converted_df = converted_df.replace(binary_dict)
```

```python
converted_df.head(10)
```

####Ordinal columns to numbers

```python
values = {'Do not know' : 0, 'I have no trust at all': 1, \
               'I have pretty little trust' : 2, \
                'I have quite a lot of trust' : 3, 'I have full trust' : 4 }

answers_list = ['How much you trust in each of these institutions: the media',
              'How much you trust in each of these institutions: the Supreme Court',
              'How much you trust in each of these institutions: the police',
              'How much you trust in each of these institutions: President of the State',
              'How much you trust in each of these institutions: the IDF',
              'How much you trust in each of these institutions: the government',
              'How much do you trust in each of these institutions: the parties',
              'How much you trust in each of these institutions: the introduction']

for column in answers_list:
    column_dict = {column : values}
    converted_df = converted_df.replace(column_dict)
```

```python
converted_df['How much you trust in each of these institutions: the media'].head(10)#check
```
